# Extracting Weather Forecast From API

```python
import requests

def api_get_data(url: str):
    response = requests.get(url, stream=True)
    response.raise_for_status()

    counter: int = 0
    limit: int = 10
    
    for items in response.json():
        yield items

if __name__ == '__main__':
    api_get_data('https://api.data.gov.my/weather/forecast')
```

```python
import dlt

url="https://api.data.gov.my/weather/forecast"

pipeline = dlt.pipeline(pipeline_name='forecast_pipeline', destination='duckdb', dataset_name='weather_data', full_refresh=True)
info = pipeline.run(api_get_data(url), table_name="forecast", write_disposition="replace")

print(info)
```

```python

import duckdb 
import pandas as pd

conn = duckdb.connect("forecast_pipeline.duckdb")

# let's see the tables
conn.sql(f"SET search_path = 'weather_data'")
print('Loaded tables: ')
display(conn.sql("show tables"))

# and the data

print("\n\n\n forecast table below:")

rides = conn.sql("SELECT * FROM forecast where location__location_id == 'Ds001'").df()
display(rides)

conn.close()
```
